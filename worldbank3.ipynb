+ ### **To perform a t-test**

```python
from scipy.stats import ttest_ind
high_income = df[df['IncomeGroup'] == 'High income']
print(high_income)
low_income = df[df['IncomeGroup'] == 'Low income']
print(low_income)
t_stat, p_value = ttest_ind(high_income['GDP per capita (USD)'], low_income['GDP per capita (USD)'])
print(t_stat, p_value)
```

+ ### **Drop rows with NaN or inf values**


```python
df = df.dropna(subset=['GDP per capita (USD)', 'Life expectancy at birth (years)'])
df = df[~df.isin([float('inf'), float('-inf')]).any(axis=1)]
```

+ ### **Pearson correlation test**

```python
from scipy.stats import pearsonr
```

```python
corr, p_value = pearsonr(df['GDP per capita (USD)'], df['Life expectancy at birth (years)'])
print(f"Pearson correlation coefficient: {corr}")
print(f"p-value:Â {p_value}")
```

+ ### **T-test**

```python
t_stat, p_value_t = ttest_ind(df['GDP per capita (USD)'], df['Life expectancy at birth (years)'])
print(f"T-test statistic: {t_stat}")
print(f"T-test p-value: {p_value_t}")
```

+ ### **Chi-square test**

```python
from scipy.stats import ttest_ind, chi2_contingency
```

```python
contingency_table = pd.crosstab(df['GDP per capita (USD)'], df['Life expectancy at birth (years)'])
chi2, p_value_chi, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-square statistic: {chi2}")
print(f"Chi-square p-value: {p_value_chi}")
print(f"Degrees of freedom:Â {dof}")
```

+ ### **Interpret the p-values and draw a conclusion**

```python
alpha = 0.05  # Significance level

if p_value_t < alpha:
    print("The t-test indicates a statistically significant difference between GDP per capita and life expectancy.")
else:
    print("The t-test does not indicate a statistically significant difference between GDP per capita and life expectancy.")

if p_value_chi < alpha:
    print("The chi-square test indicates a statistically significant association between GDP per capita and life expectancy.")
else:
    print("The chi-square test does not indicate a statistically significant association between GDP per capita and lifeÂ expectancy.")
```
