```python
import pandas as pd

df = pd.read_excel('WorldBank.xlsx')
df
```

```python
europe_central_asia = df[df['Region'] == 'Europe & Central Asia']
europe_central_asia
```

```python
birth_rate_by_income = df.groupby('IncomeGroup')['Birth rate, crude (per 1,000 people)'].agg(['mean', 'count'])
birth_rate_by_income
```

```python
mean_gdp_per_capita = df['GDP per capita (USD)'].mean()
print(mean_gdp_per_capita)
median_gdp_per_capita = df['GDP per capita (USD)'].median()
print(median_gdp_per_capita)
std_gdp_per_capita = df['GDP per capita (USD)'].std()
print(std_gdp_per_capita)
```

```python
import matplotlib.pyplot as plt
df.boxplot(column='GDP per capita (USD)')
plt.show()
```

```python
df['GDP per capita (USD)'].hist()
plt.show()
```

```python
skewness = df['GDP per capita (USD)'].skew()
print(skewness)
kurtosis = df['GDP per capita (USD)'].kurt()
print(kurtosis)
```

```python
from scipy.stats import ttest_ind
high_income = df[df['IncomeGroup'] == 'High income']
print(high_income)
low_income = df[df['IncomeGroup'] == 'Low income']
print(low_income)
t_stat, p_value = ttest_ind(high_income['GDP per capita (USD)'], low_income['GDP per capita (USD)'])
print(t_stat, p_value)
```
